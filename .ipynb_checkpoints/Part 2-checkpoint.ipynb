```python
import pandas as pd
data = pd.read_csv('day1prediction.csv')
```

```python
data.head()
```

### K-Nearest Neighbors Classifier

```python
from sklearn.model_selection import train_test_split
import numpy as np
train, test = train_test_split(data, test_size=0.2)
x_train = train.loc[:,'Close_DJI':'XAU_SNP']
y_train = train.loc[:,'DJI_next_day':'SNP_next_day']

x_test = test.loc[:,'Close_DJI':'XAU_SNP']
y_test = test.loc[:,'DJI_next_day':'SNP_next_day']

from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
neigh.predict(x_test)
```

```python
y_pred = neigh.predict(x_test)

accuracy = [0]*5

for i in range(len(y_test)):
    for j in range(5):
        if y_test[i][j] == y_pred[i][j]:
            accuracy[j]+=1

accuracy = [(x*1.0/len(y_test))*100 for x in accuracy]
print(accuracy)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree_clf = DecisionTreeClassifier(random_state=0)
dec_tree_clf.fit(x_train, y_train)
y_pred_dt = dec_tree_clf.predict(x_test)

accuracy = [0]*5

for i in range(len(y_test)):
    for j in range(5):
        if y_test[i][j] == y_pred_dt[i][j]:
            accuracy[j]+=1

accuracy = [(x*1.0/len(y_test))*100 for x in accuracy]
print(accuracy)
```
