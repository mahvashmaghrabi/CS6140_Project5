```python
import pandas as pd
data_DJI = pd.read_csv('CNNpred/Processed_DJI.csv')
data_NASDAQ = pd.read_csv('CNNpred/Processed_NASDAQ.csv')
data_NYSE = pd.read_csv('CNNpred/Processed_NYSE.csv')
data_RUSSELL = pd.read_csv('CNNpred/Processed_RUSSELL.csv')
data_SnP = pd.read_csv('CNNpred/Processed_S&P.csv')
```

```python
data_DJI.head()
```

```python
data_NASDAQ.head()
```

```python
output4 = pd.merge(data_DJI, data_NASDAQ,
                   on='Date', 
                   how='outer')

output4.head()
```

```python
output5 = pd.merge(output4, data_NYSE,
                   on='Date', 
                   how='outer')

output5.head()
```

```python
output6 = pd.merge(output5, data_RUSSELL,
                   on='Date', 
                   how='outer')

output6.head()
```

```python
output7 = pd.merge(output6, data_SnP,
                   on='Date', 
                   how='outer')

output7.head()
```

```python
output7 = output7.fillna(0)
```

```python
output7.to_csv('day1prediction.csv', sep='\t', encoding='utf-8')
```
